# Vetores como dados

## Ideia central


- Um vetor representa uma observa√ß√£o
- Cada componente representa uma vari√°vel (feature)

## Exemplos conceituais

- Pessoa -> (idade, altura)
- Produto -> (pre√ßo, avalia√ß√£o, vendas)

## Criando um vetor com NumPy

```python
import numpy as np

pessoa = np.array([25, 1.75])

pessoa
```

```python
# V√°rios vetores (matriz) de dados

dados = np.array([
    [30, 1.77],
    [17, 1.83],
    [24, 1.65]
])

dados
```

```python
# Transformando isso num DF do Pandas

import pandas as pd

df = pd.DataFrame(dados, columns=["idade", "altura"])

df

# Cada linha do DF √© um vetor
# Cada elemento da linha √© uma feature daquele vetor
# Dataset = Matriz
```
